# DIA Price Visualization

This notebook visualizes the data collected by the Java price tracker in [`src/App.java`](src/App.java).

**How it fits together**

- `src/App.java` polls the Twelve Data API for the `DIA` price every 15 seconds and appends each result as one line of JSON to `data/prices.jsonl` (see "Storage" in the [README](README.md)).
- This notebook only reads that file — it does not install, compile, or run any Java. Start the tracker separately first (see "Running locally" in the README), let it collect a few data points, then run the cells below.
- Because the notebook just reads a snapshot of the file, re-run the cells any time (the tracker can keep running in the background) to refresh the chart with newly collected points.

**Requirements:** `pandas` and `matplotlib` (`pip install pandas matplotlib`).

## Load the collected data

Each line in `data/prices.jsonl` is a standalone JSON object, e.g. `{"price":420.15,"timestamp":"2026-07-19T12:34:56.789Z"}`. We read the file line by line and load it into a DataFrame.

```python
import json
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

DATA_FILE = Path("data/prices.jsonl")
```

```python
records = []
with DATA_FILE.open() as f:
    for line in f:
        line = line.strip()
        if line:
            records.append(json.loads(line))

df = pd.DataFrame(records)
df["timestamp"] = pd.to_datetime(df["timestamp"])
df
```

## Plot price over time

```python
plt.figure(figsize=(10, 5))
plt.plot(df["timestamp"], df["price"], marker="o")

plt.title("DIA Price Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Price (USD)")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
